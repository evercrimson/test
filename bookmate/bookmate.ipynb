```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
data = pd.read_csv("analyst_test_transactions_db.csv")
```

```python
data.sample(5)
```

```python
data.info()
```

Тип order_id сменился на float, вернём его в нормальное состояние.

```python
data.order_id = data.order_id.astype(int);
```

Заметим, что в базе данных есть мусорные значения, нам необходимо удалить их.

```python
data[data['user_id'] < 0].sample(5)
```

```python
data = data.query("user_id > 0 and order_id > 0")
```

```python
data['transaction_date'] = pd.to_datetime(data['transaction_date'], errors='coerce')
data['reg_date'] = pd.to_datetime(data['reg_date'], errors='coerce')
data = data.query("reg_date != '1900-01'")
```

Т.к. анализ проводим по месяцам, дни можно отбросить

```python
data['reg_date'] = data['reg_date'].dt.strftime('%Y-%m')
data['transaction_date'] = data['transaction_date'].dt.strftime('%Y-%m')
data.sample(5)
```

Для начала просто посмотрим, сколько транзакций совершили пользователи, зарегестрированные в определённый месяц.

```python
trncount = data.groupby('reg_date')['order_id'].nunique()
trncount.tail(18)[:-2]
```

Видно, что пользователи зарегестрировавшиеся в июле 2015-го совершают много транзакций по сравнению со своими "соседями", при том что год назад в этом месяце наблюдалась пониженная активность, т.е. сезонность роли не играет. Это первое свидетельство в пользу успешности акции, однако следует провести и дополнительный анализ.
Для наглядности построим график.

```python
plt.figure(figsize=(19,6))
ax = plt.subplot(1, 1, 1)
ax.plot(trncount.index[1:], trncount[1:]) # для удобства откинем первый месяц
xticks = ax.get_xticks()
ax.set_xticks(xticks[::5])
ax.set_xlabel('Месяц регистрации', fontsize=14)
ax.set_ylabel('Количество совершённых транзакций', fontsize=14)
plt.grid(color='black', linestyle='--', linewidth=1)
plt.show()
```

Даже несмотря на то, что у поздних пользователей было мало времени на совершение транзакций, в июле почти был достигнут глобальный максимум.

Когортный анализ
----------------------------

```python
grouped = data.groupby(['reg_date', 'transaction_date'])
cohorts = grouped.agg({'user_id' : pd.Series.nunique, 'order_id' : pd.Series.nunique})
cohorts.rename(columns = {'user_id' : 'total_users', 'order_id' : 'total_orders'}, inplace = True)
cohorts.tail(10)
```

```python
def cohort_period(d):
    d['month'] = np.arange(len(d)) + 1
    return d

cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts.tail(6)
```

Теперь для каждой когортной группы посчитаем retention по месяцам.

```python
cohorts.reset_index(inplace=True)
cohorts.set_index(['reg_date', 'month'], inplace=True)
```

Нам неизвестно точное количество зарегестрированных пользователей за месяц, поэтому придётся за него принять число зафиксированных уникальных user_id. Таким образом, мы будем учитывать только пользователей, совершивших хотя бы одну транзакцию, что снижает качество анализа.

```python
cohort_group_size = data.groupby('reg_date')['user_id'].nunique()
cohort_group_size.tail()
```

```python
user_retention = cohorts['total_users'].unstack(0).divide(cohort_group_size, axis=1)
user_retention.head(10)
```

```python
user_retention[['2015-04', '2015-05', '2015-06', '2015-07', '2015-08']].plot(figsize=(12,6))
plt.title('Retention пользователей')
plt.xticks(np.arange(1, 12.1, 1))
plt.xlim(1, 12)
plt.ylabel('% пользователей совершивших транзакцию');
```

Видно, что в результате акции резко возрос retention на второй месяц. Для более поздних месяцев данных пока недостаточно и большая часть пользователей зафиксирована в дату регистрации. Чтобы оценить долгосрочные последствия акции стоит накопить данные ещё на пару месяцев вперёд. Однако, пока можно сказать что акция была успешна.
